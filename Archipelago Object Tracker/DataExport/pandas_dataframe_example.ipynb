```python
import pandas as pd

# create data
data = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}

# load data into a DataFrame object:
df = pd.DataFrame(data, index = ["day1", "day2", "day3"])

# print DataFrame table
print("Print DataFrae Table")
print(df)
print('\n')

# refer to the row index
print("Print selected row")
print(df.loc["day1"])
print('\n')

# use a list of indexes
print("Print selected rows")
print(df.loc[["day1", "day2"]])
print('\n')

# print whole column
print("Print selected column")
print(df['calories'])
print('\n')

# create new data frame with rename column and index row
df_new = df.rename(columns={'calories': 'Cal'}, index={'day1': 1})
print("new data frame")
print(df_new)
print("original data frame")
print(df)
print('\n')

# change original data frame with parameter inplace
df.rename(columns={'calories': 'Cal'}, index={'day1': 1}, inplace=True)
print(df)
print('\n')

# add new column 
exercise = ['Running', 'Walking', 'Running']
df["exercise"] = exercise

# using insert
df.insert(2, "km", [3,2,3], True)
print(df)
print('\n')

# add new row to the data frame
df.loc["day4"] = {'Cal':'400', 'duration':40, 'km':2, 'exercise':'Walking'}
print(df)
print('\n')

# combine two data frames
df1 = pd.DataFrame(data, index = ["day1", "day2", "day3"])
df2 = pd.DataFrame(data, index = ["day4", "day5", "day6"])
com_df = pd.concat([df1,df2])
print(com_df)
print('\n')
```
